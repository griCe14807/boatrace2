```python
import glob
import os

race_file_set = set([os.path.basename(file) for file in glob.glob(os.path.join('../../data/crawledData', '*.pkl'))])
start_time_file_set = set([os.path.basename(file) for file in glob.glob(os.path.join('../../data/crawledData/start_time', '*.pkl'))])

merged_file_set = set([os.path.basename(file) for file in glob.glob(os.path.join('../../data/crawledData/st_merged', '*.pkl'))])

common_file_name_list = list(race_file_set & start_time_file_set - merged_file_set)
common_file_name_list
```

```python
import pandas as pd

for filename in common_file_name_list:
    start_time_df = pd.read_pickle('../../data/crawledData/start_time/{0}'.format(filename))
    
    race_df = pd.read_pickle('../../data/crawledData/{0}'.format(filename))
    
    merged_df = pd.merge(race_df, start_time_df, how="left", left_index=True, right_index=True)
    
    merged_df.to_pickle('../../data/crawledData/st_merged/{0}'.format(filename))
```
