```python
import glob
import os

race_file_set = set([os.path.basename(file) for file in glob.glob(os.path.join('../../../data/crawledData', '*.pkl'))])
start_time_file_set = set([os.path.basename(file) for file in glob.glob(os.path.join('../../../data/crawledData/exhibition_mod', '*.pkl'))])

merged_file_set = set([os.path.basename(file) for file in glob.glob(os.path.join('../../../data/crawledData/exhibition_mod_merged', '*.pkl'))])

common_file_name_list = list(race_file_set & start_time_file_set - merged_file_set)
print("{0}ファイルを統合します".format(len(common_file_name_list)))
```

```python
import pandas as pd

for filename in common_file_name_list:
    exhibition_mod_df = pd.read_pickle('../../../data/crawledData/exhibition_mod/{0}'.format(filename))
    
    race_df = pd.read_pickle('../../../data/crawledData/{0}'.format(filename))
    race_df.drop(["exhibition_cource", "exhibition_ST", "flying", "late"], axis=1, inplace=True)
    
    merged_df = pd.merge(race_df, exhibition_mod_df, how="left", left_index=True, right_index=True)
    
    merged_df.to_pickle('../../../data/crawledData/exhibition_mod_merged/{0}'.format(filename))
```
